```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#raead the data
iris_data=pd.read_csv("C:/Users/Admin/Desktop/R prog for DS/Iris.csv")
iris_data.head()
```

```python
#Aim is to create a decision tree classifier and visualise graphically
irisdf=pd.DataFrame(iris_data,columns=iris_data.columns)
irisdf.head()
```

```python
X=iris_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=iris_data[['Species']]
```

```python
#Splitting data into train and test set
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
#fit the model to given data
model.fit(x_train,y_train)
```

```python
#plot the decision tree
from sklearn.tree import plot_tree
plt.figure(figsize = (20,10))
plot_tree(model,filled = True)
plt.show()
```

```python
#accuracy of the model
model.score(x_test,y_test)
```
